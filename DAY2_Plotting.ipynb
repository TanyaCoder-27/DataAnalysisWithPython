```python
#There are 5 subjects: Psychology,Nanotechnology,Medicine,Oncology,Anthropology. 2 Students have applied to do masters in any of these 5 fields.
#with a line plot,show in which field they can excel?The cutoff for each subject is 70.
#Student1=[67,45,73,71,53]
#Student2=[43,67,68,53,85]
Sub=['Psychology','Nanotechnology','Medicine','Oncology','Anthropology']
cutoff=[70,70,70,70,70]
plt.plot(Sub,cutoff,label="Cutoff",color='g',marker='P',linestyle=":",linewidth=4)#linestyle and linewidth are for lines only
#plt.plottype(X,Y,color,marker,linewidth,linestyle)
S1=[67,45,73,71,53]
S2=[43,67,68,53,85]
plt.scatter(Sub,S1,label="Student1",color='b',marker='p')#scatter for dots in place of line
plt.scatter(Sub,S2,label="Student2",color='k',marker='H')
plt.title("2nd plot")
plt.xlabel('Subjects')
plt.ylabel("Marks")
plt.legend()#for label seen

```

```python
#An Ind vs Aus match is on live.India has given 298 as target for Aus. The runs for every 5hrs is given as follows:
#Ind=[25,51,84,131,160,189,220,250,267,297]
#Aus completed only upto 25 overs.
#Aus=[15,41,94,110,151]
#plot overs vs runs for both the countries with India as a blue line and aus as yellow line.
#give appropriate labels and legend. the marker for india is "flag" and aus is "plus"
#take overs from arange function and create runs as arrays.Take linewidth=2.
import numpy as np
overs1=np.arange(5,51,5)
overs2=np.arange(5,26,5)
Ind=np.array([25,51,84,131,160,189,220,250,267,297])
Aus=np.array([15,41,94,110,151])
plt.plot(overs1,Ind,label="India",color='b',marker=">",linewidth=2)
plt.plot(overs2,Aus,label="Australia",color='y',marker="P",linewidth=2)
plt.legend()
plt.title("IND VS AUS")
plt.xlabel('Overs')
plt.ylabel("Scores")
plt.grid()#for a grid visibility

```
