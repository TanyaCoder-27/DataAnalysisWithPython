```python
#Take a number from user which is his salary. If the salary is more than 50,000, there is a 10% tax(gets cut). Print his final salary.
#The default datatype of input is String.
sal = int(input("enter salary:"))
if( sal>=50000):
  tax = (10/100)*sal
  sal=sal-tax
  print("After tax deduction, your salary is:",sal)
else:
  print("No tax deduction. your salary is:",sal)
```

```python
#Get the salary from the user, if the salary is more than or equal to 50000, apply 10% tax deduction.
#if the salary is less than 50000, apply 5% tax. if the salary is more than 70000,apply 12% tax. If your salary exceeds 1lakh, print "you are promoted to next level"
sal=int(input("Enter Salary:"))
if sal>100000:
  print("You are Promoted")
elif sal>70000:
  tax=(12/100)*sal
  print("salary is",sal-tax)
elif sal>50000:
  tax=(10/100)*sal
  print("salary is",sal-tax)
else:
  tax=(5/100)*sal
  print("salary is",sal-tax)
```

```python
#list_name=[output for if]
l=[24,56,13,24,57,10,42,33,6,9]
res=[i*i for i in l if i>=50]
print(res)
```

```python
#append if its greater than 50 ELSE X
#newlist=[(stat of if) if(cond) else (stat of else) for]
l=[24,56,13,24,57,10,42,33,6,9]
res=[i*i  if i>=50 else'X'for i in l]
print(res)
```
