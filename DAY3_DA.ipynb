```python
df=pd.read_csv("/content/grades_withnulls.csv")
df
```

```python
#Data cleaning - deealing with missing values
#1st - know where there are NaN(Not A Number) values by isnull() function

print(df.isnull())
```

```python
#.sum() along with isnull() function provides col number too
print(df.isnull().sum())
```

```python
# Handling NaN -- 1.remove 2.replace
#1.remove -dropna()
#dropna() - just drops the NaN element but doesn't save it
#dropna(inplace="True")- to save changes in a dataframe
print(df.dropna())
```

```python
df #didn't show the dropped dataframe
```

```python
df2=df.dropna(inplace=True)
df2
```

```python
df#now the dataframe is changed
```

```python
#2.replace - fillna()
```

```python
df3=pd.read_csv("/content/grades_withnulls.csv")
df3.fillna(555,inplace=True) #fillna(value) - notchanged(not worked), #fillna(value,inplace=True) - changed!
df3
```

```python
df4=pd.read_csv("/content/grades_withnulls.csv")
mv=df4['SEM1'].mean() #mean value of col 'SEM1' is calculated and stored in mv
print(mv)
print("--------------------------------------")
df4.fillna(mv,inplace=True)#replacing the NaN values with mean value here
df4
```
