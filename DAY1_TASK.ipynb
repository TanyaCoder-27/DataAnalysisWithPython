```python
#A tuple is a collection which is ordered and unchangeable and allow duplicates.
#Tuples are written with round brackets and are indexed.
#Create
t=('apple',73,65.5,'cherry','apple')
print(t)
```

```python
#Print the number of items in the tuple.
print(len(t))
```

```python
#One item tuple, remember the comma.

t1 = ("apple",)
print(type(t1))

#NOT a tuple
t2 = ("apple")
print(type(t2))
```

```python
#access tuple items by referring to the index number, inside square brackets.
tup = (23,45,"meow",606.7)
print(tup[2])
print(tup[1:3]) #range of indexes
for i in tup:
  print(i)
```

```python
#immutable. Can't modify values once initialized.So, to change any value, convert the tuple into list.
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)
print(x)
```

```python
#For any updation like appending,removing tuples are converted into lists initially
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
print(thistuple)

y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)
print(thistuple)
```

```python

# Add tuple to a tuple.
x = ("apple", "banana", "cherry")
y = ("orange",)
x += y

print(x)
```

```python
#The del keyword can delete the tuple completely
a=(12,543,98.99)
print(a)
del a
print(a) #this will raise an error because the tuple no longer exists
```

```python
#joining & Multiplying two tuples
tp1=(12,56,78)
tp2=(30,76,456)
tp1=tp1+tp2
print(tp1)
print(tp2*2)
```

```python
#A set is a collection which is unordered,unchangeable,unindexed and do not allow duplicate values.
#Create a Set
s = {"apple",345,89.7, "banana",True, "cherry",False,0} #False and 0 is considered the same value
print(s)
s2=set(('meow',333,"tom",77.77))
print(s2)
```

```python
#length of set
print(len(s))
```

```python
#type
print(type(s))
```

```python
#Accessing
for i in s:
  print(i)
```

```python
#check for an element
print("tom" in s2)
print("tom" in s)
```

```python
#Once a set is created, you cannot change its items, but you can add new items by using add() method.
s2.add("jerry")
print(s2)
```

```python
#adding sets by update() method
s.update(s2)
print(s)
l=[23,"dog",789.6]
s2.update(l)
print(s2)
```

```python
#Removing through remove() and discard()
s.remove("jerry")
s2.remove("jerry")
print(s)
print(s2)
s2.remove("duck") #If the item to remove does not exist, remove() will raise an error.
```

```python
s2.discard("bat")
s.discard(True)
print(s2) #If the item to remove does not exist, discard() will NOT raise an error.
print(s)
#Sets are unordered, so when using the pop() method, you do not know which item that gets removed.
```

```python


d2={999,54,5,"hey"}
del d2 #raises error as the set got completely deleted

print(d2)
```

```python
d={34,56,78}
d.clear() #no error. just empties the set
print(d)
```

```python
#Joining two sets
#The union() method returns a new set with all items from both sets
s1={'a','b','c',4}
s2={1,2,3,4}
s1=s1.union(s2)
print(s1)
#The update() method inserts the items in set2 into set1
s1.update(s2)
print(s1)
#Both union() and update() will exclude any duplicate items
```

```python
#Set Methods
#The intersection() method will return a new set, that only contains the items that are present in both sets.
a={1,2,3,4,5}
v={5,4,7,8}
print(a.intersection(v))
#The symmetric_difference() method will return a new set, that contains only the elements that are NOT present in both sets.
print(a.symmetric_difference(v))
```
